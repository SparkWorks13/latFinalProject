## The Metric
   Our metric was the neighborhood that is the most fun. To determine this, we selected a few types of "attractions" and measured their quantities in neighborhoods around Pittsburgh. The idea was that the most "fun" neighborhood would have the most instances of these selected attractions. We selected the following datasets based on this idea:
 - Ballfields
 - FishFry
 - POGOHPARKs
 - Pools
 
Each of these datasets contains a wealth of information on their respective topics (given by their names) such as the duration they have remained open, owners (if applicable), etc. For us, the piece of information we were looking for was their geological coordinates, which we could use to determine their neighborhood of operation.
 
## The Best Neighborhood
   The advantage to dividing our data in this way is that there is also a clearly defined "best" neighborhood for each of these submetrics; i.e, the neighborhood with the most instances of the given attraction.
   - The best neighborhoos for ballfields were South Side Slopes and Brighton Hills, both with 7 ballfields
   - The best neighborhood for fish fry vendors was Central Business District, with 5
   - The best neighborhood for POGOPARKS was Central Oakland, with 7 instances
   - The best neighborhood for pools was another tie between Highland Park and South Side Flats, both with 2.
   
To determine the best overall neighborhood, we simply summed the total attractions in each neighborhood, and came up with two frontrunners: Central Business District and South Side Flats, both with 11 total instances of "fun" attractions. To determine which, between these two, was the more fun neighborhood, we compared their distribution of attractions; whichever was more variable would be deemed the winner. By this metric, South Side Flats won, as Central Business District had no instances of either ballfields nor pools, whilst South Side Flats had a fairly even distribution of all submetrics we selected.

```python
import pandas as pd
import fpsnippets as fp
```

```python
fish = pd.read_csv("datasets/FishFry.csv")
```

```python
fish.head(10)
```

```python
ll = fish.iloc[:, 20:]
neighborhoodsF = {}
for x in range(ll.iloc[:, 0].count()):
    loc = fp.geo_to_neighborhood(ll.iloc[x, 0], ll.iloc[x, 1])
    if loc in neighborhoodsF:
        neighborhoodsF[loc] += 1
    elif loc != None:
        neighborhoodsF[loc] = 1
```

```python
print(neighborhoodsF)
```

```python
n = pd.DataFrame(neighborhoodsF.items())
```

```python
n.plot.bar(x=0)
```

This bar graph shows that the Central Business District has the most fish fry locations. Most of the other neighborhoods have 1-2 locations.

```python
# flat sum "objective" fun
neighborhoods = {i: neighborhoodsF.get(i, 0) + neighborhoodsPG.get(i, 0) + neighborhoodsB.get(i, 0) + neighborhoodsP.get(i, 0) for i in set(neighborhoodsF).union(neighborhoodsPG).union(neighborhoodsB).union(neighborhoodsP)}
```

```python
print(neighborhoods)
```

```python
n = pd.DataFrame(neighborhoods.items())
```
